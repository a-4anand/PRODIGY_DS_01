```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

READING THE CSV FILE

```python
df=pd.read_csv("task1Data.csv", encoding="utf-8")
```

```python
df.head()
```

```python

df_final = df.drop(['Country Code',	'Indicator Name',	'Indicator Code'], axis=1)
```

```python
df_final.tail()
```

```python
Country_Data = df_final[df_final["Country Name"] == "India"]
```

```python
Country_Data
```

```python
# Filter out non-numeric column names
years = [int(column) for column in Country_Data.columns[2:] if column.isdigit()]

# Extract population data
population = Country_Data.iloc[:, 2:].values.flatten()

# Plotting code goes here
plt.figure(figsize=(5, 3))


plt.bar(years, population[:-1], color='skyblue') # since data for 2022 is missig we are not considering thata
plt.xlabel('Year')
plt.ylabel('Population')
plt.title(f'Population of India Over the Years')

plt.show()
```
